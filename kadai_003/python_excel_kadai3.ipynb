```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill

file_list = ["2022_年間売上表.xlsx", "2023_年間売上表.xlsx"]
df = pd.DataFrame()

for file_name in file_list:
    temp_df = pd.read_excel(file_name)
    temp_df.columns = ["売上年", "商品","金額(千円)"]
    df = pd.concat([df, temp_df], ignore_index = True)
df = df[["商品", "売上年", "金額(千円)"]]

result = df.groupby(["商品","売上年"])["金額(千円)"].sum().reset_index()

with pd.ExcelWriter("売上集計表.xlsx") as writer:
    result.to_excel(writer, index = False)



wb = openpyxl.load_workbook("売上集計表.xlsx")
ws = wb.active

color = PatternFill(patternType='solid', fgColor='F2F2F2')

for col in ws['A1:C1']:
    for cell in col:
        cell.fill = color

wb.save("売上集計表.xlsx")
```
