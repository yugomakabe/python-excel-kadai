```python
import pandas as pd

data = {
    "日付" : ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    "社員名" : ['山田', '佐藤', '鈴木', '田中', '高橋'],
    "売上" : [100, 200, 150, 300, 250],
    "部門" : ['メーカー', '代理店', 'メーカー', '商社', '代理店']
}

df = pd.DataFrame(data)

df["平均売上"] = df["売上"].mean()

def performance(rank):
    result = ''
    ave = df["売上"].mean()
    if rank >= ave + 50:
        result = 'A'
    elif rank >= ave:
        result = 'B'
    else:
        result = 'C'
    return result
df["業績ランク"] = df["売上"].apply(performance)

with pd.ExcelWriter("業績.xlsx") as writer:
    df.to_excel(writer, sheet_name = '業績', index = False)
```
