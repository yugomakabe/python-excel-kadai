```python
import openpyxl
from openpyxl.styles import Side, Border
from datetime import datetime

wb = openpyxl.Workbook()
ws = wb.active

ws['A1'].value = "請求書"
ws['A2'].value = "株式会社ABC"
ws['A3'].value = "〒101-0022 東京都千代田区神田練塀町300"
ws['A4'].value = "TEL:03-1234-5678 FAX:03-1234-5678"
ws['A5'].value = "担当者名:鈴木一郎 様"
ws['E2'].value = "No."
ws['E3'].value = "日付"
ws['F2'].value = "0001"

today = datetime.today()
ws['F3'].value = today.strftime("%Y/%m/%d")

header = ["商品名", "数量", "単価", "金額" ]
ws.append(header)

data = [["商品A", 2, 10000], ["商品B", 1, 15000]]
for row in data:
    name, qty, price = row
    total = qty * price
    ws.append([price, qty, price, total])

ws['D9'].value = "=E11+E12 "

ws['A10'].value = "合計"
ws['A11'].value = "消費税"
ws["A12"].value = "税込合計"

ws['D10'].value = "=E13"
ws['D11'].value = "=E15*0.1"
ws['D12'].value = "=E15+E16"

thin_border = Side(style = 'thin')
border = Border(top = thin_border, bottom = thin_border, left = thin_border, right = thin_border )
for row in ws['A6:D9']:
    for cell in row:
        cell.border = border
for row in ws['A10:D12']:
     for cell in row:
         cell.border = border

ws.insert_rows(1, 1)
ws.insert_rows(3, 1)
ws.insert_rows(8, 2)
ws.insert_rows(14, 1)
ws.insert_cols(1, 1)

wb.save(f"請求書_{today.strftime("%Y%m%d")}.xlsx")
```
